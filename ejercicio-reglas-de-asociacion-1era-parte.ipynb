2) Cuál es el soporte de cada 1-itemset?

```python
for columna in df.columns:
  print("El itemset {"+columna+"} tiene un soporte de "+str(sum(df[columna])/len(df)))
```

3) Cuál es el ancho promedio de cada transacción? 

```python
acumulador = 0
for index, row in df.iterrows():
  acumulador = acumulador + row.sum()

print("El ancho promedio de una transacción es: "+str(acumulador/len(df)))
```

4) Cuál es la transacción más larga?  

```python
maximo = (0,-1)
for index, row in df.iterrows():
  if row.sum() > maximo[0]:
    maximo = (row.sum(),index)

print("La transacción más larga es: "+str(maximo[1]))
```

6) Cuál es el soporte de la regla AB->C 

```python
print("El itemset {ABC} y la regla AB -> C tienen un soporte de "+str(len(df[df['A'] == 1][df['B'] == 1][df['C'] == 1])/len(df)))
```

7) Cuál es la confianza de la regla del punto anterior? 

```python
soporteABC = len(df[df['A'] == 1][df['B'] == 1][df['C'] == 1])/len(df) 
soporteAB = len(df[df['A'] == 1][df['B'] == 1])/len(df)
print("La confianza de la regla AB -> C es: "+str(soporteABC/soporteAB))
```

8) Y el lift?  

```python
soporteABC = len(df[df['A'] == 1][df['B'] == 1][df['C'] == 1])/len(df)
soporteAB = len(df[df['A'] == 1][df['B'] == 1])/len(df)
soporteC = len(df[df['C'] == 1])/len(df)
print("El lift de la regla AB -> C es: "+str(soporteABC/(soporteAB*soporteC)))
```
